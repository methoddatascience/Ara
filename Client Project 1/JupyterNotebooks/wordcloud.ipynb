# Generate Wordcloud from Weighted (TF-IDF) Pubmed Corpus

```python
import numpy as np
import matplotlib.pyplot as plt
from os import path
from wordcloud import WordCloud
import pandas as pd
import os
from PIL import Image
import PIL.ImageOps

#load in and pre-process dataset
text = pd.read_csv("mean_tfidf_corpus.csv").sort_values(by='tfidf',ascending=False).drop("Unnamed: 0",axis=1)
text = text.set_index('variable')

# convert to dictionary
dic = text.to_dict()

#load in mask, comes from https://www.iconfinder.com/icons/1094099/cross_health_hospital_medicine_sign_icon
cross_mask = np.array(Image.open("cross.png"))

wc = WordCloud(background_color="white",mask=cross_mask,random_state=6,max_words=2000).fit_words(dic['tfidf'])

plt.figure(figsize=(16,10))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
```
